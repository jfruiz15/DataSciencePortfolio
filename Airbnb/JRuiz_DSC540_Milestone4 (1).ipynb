```python
# importing libraries
import urllib.request, urllib.parse
from urllib.error import HTTPError,URLError
import json
import pandas as pd
import requests
```

```python
url = 'https://data.ny.gov/resource/i9wp-a4ja.json'
uh = urllib.request.urlopen(url)
resp = requests.get(url)
resp
```

```python
data = resp.json()
```

```python
subways = pd.DataFrame(data)
```
