```python
#Import libraries
import requests
import html5lib
from bs4 import BeautifulSoup
```

```python
import pandas as pd
import numpy as np
import csv
```

```python
url = "https://en.wikipedia.org/wiki/Tourism_in_New_York_City"
response = requests.get(url)
print(response.status_code)
soup = BeautifulSoup(response.text,"html.parser")
table = soup.findAll('table',{"class":"wikitable"})[1]
tr = table.findAll(['tr'])[1:20]
csvFile = open("empirestateofmind.csv",'wt',newline='', encoding='utf-8')
writer = csv.writer(csvFile)  
try:   
        for cell in tr:
            th = cell.find_all('th')
            th_data = [col.text.strip('\n') for col in th]
            td = cell.find_all('td')
            row = [i.text.replace('\n','') for i in td]
            writer.writerow(th_data+row)      
        
finally:   
    csvFile.close()
```
